```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import roc_auc_score

import sklearn.metrics
from sklearn.cross_validation import train_test_split, cross_val_score
```

```python
boston = datasets.load_boston()
```

```python
boston.keys()
```

```python
features = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
features.head()
```

```python
features.drop(['B'],axis=1, inplace=True)
```

```python
features.head()
```

```python
targets = boston.target
```

```python
targets
```

```python
np.isnan(features['LSTAT']).any()
```

```python
X=features
```

```python
y=targets
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=0)
```

```python
X_train
```

```python
clf = DecisionTreeRegressor(random_state=1, min_samples_leaf=2)
clf.fit(X_train, y_train)
```

```python
clf2 = DecisionTreeRegressor(random_state=1, max_depth=5)
clf2.fit(X_train, y_train)

```

```python
predictions = clf.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,predictions)
```

```python
predictions2 = clf2.predict(X_test)
```

```python
r2_score(y_test,predictions2)
```

```python
features.drop(['ZN'],axis=1, inplace=True)
features.drop(['INDUS'],axis=1, inplace=True)
features.drop(['CHAS'],axis=1, inplace=True)
```

```python
features.drop(['NOX'],axis=1, inplace=True)
features.drop(['RAD'],axis=1, inplace=True)
```

```python
features.head()
```

```python
X=features
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=0)
clf = DecisionTreeRegressor(random_state=1, min_samples_leaf=2)
clf.fit(X_train, y_train)
```

```python
clf2 = DecisionTreeRegressor(random_state=1, max_depth=5)
clf2.fit(X_train, y_train)
```

```python
predictions = clf.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,predictions)
```

```python
predictions2 = clf2.predict(X_test)
r2_score(y_test,predictions2)
```

```python
combined = (predictions + predictions2)/2
rounded = np.round(combined)
r2_score(y_test,rounded)
#Combined result of 2 Trees
```
